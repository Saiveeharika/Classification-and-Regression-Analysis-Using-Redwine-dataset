# Train Test Split - 80/20

```python
#Linear Regression
print(f'Linear Regression')
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
linreg_pred = linreg.predict(X_test)
mse = mean_squared_error(Y_test,linreg_pred)
print(f'Mean Squared Error of Linear : {mse}')
mae = mean_absolute_error(Y_test,linreg_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, linreg_pred)
print(f'R-squared: {r2:4f}')



#RANDOM FOREST

print(f'Random Forest')
from sklearn.ensemble import RandomForestRegressor
rforest = RandomForestRegressor(n_estimators=100,random_state=42)
rforest.fit(X_train,Y_train)
rforest_pred=rforest.predict(X_test)
mse = mean_squared_error(Y_test,rforest_pred)
print(f'Mean Squared Error : {mse}')
mae = mean_absolute_error(Y_test,rforest_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, rforest_pred)
print(f'R-squared Random : {r2:4f}')


#SVR
print(f'SVR')
from sklearn.svm import SVR
svr = SVR(kernel='linear')
svr.fit(X_train,Y_train)
svr_pred=svr.predict(X_test)
mse = mean_squared_error(Y_test,svr_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,svr_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, svr_pred)
print(f'R-squared: {r2:4f}')

#KNN

print(f'KNN')
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=30)
knn.fit(X_train,Y_train)
knn_pred=knn.predict(X_test)
mse = mean_squared_error(Y_test,knn_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,knn_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, knn_pred)
print(f'R-squared: {r2:4f}')

#Naive bayes
print(f'Naive Bayes')
from sklearn.naive_bayes import GaussianNB
naivebayes = GaussianNB()
naivebayes.fit(X_train,Y_train)
naivebayes_pred=naivebayes.predict(X_test)
mse = mean_squared_error(Y_test,naivebayes_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,naivebayes_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, naivebayes_pred)
print(f'R-squared: {r2:4f}')

#Decision Tree
print(f'Decision Tree')
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(max_depth=3)
dtree.fit(X_train,Y_train)
dtree_pred=dtree.predict(X_test)
mse = mean_squared_error(Y_test,dtree_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,dtree_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, dtree_pred)
print(f'R-squared: {r2:4f}')

#Gradient Boosting
print(f'Gradient Boosting')
from sklearn.ensemble import GradientBoostingRegressor
gradient_boosting = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gradient_boosting.fit(X_train,Y_train)
gradient_boosting_pred=gradient_boosting.predict(X_test)
mse = mean_squared_error(Y_test,gradient_boosting_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,gradient_boosting_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, gradient_boosting_pred)
print(f'R-squared: {r2:4f}')

#AdaBoostRegressor
print(f'ADA Boost')
from sklearn.ensemble import AdaBoostRegressor
base_classifier = DecisionTreeRegressor(max_depth=2)
adaboost = AdaBoostRegressor(n_estimators=50, random_state=42)
adaboost.fit(X_train,Y_train)
adaboost_pred=adaboost.predict(X_test)
mse = mean_squared_error(Y_test,adaboost_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,adaboost_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, adaboost_pred)
print(f'R-squared: {r2:4f}')

#XGBoost
print(f'XG Boost')
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train, Y_train, verbose=False)
xgb_pred=xgb.predict(X_test)
mse = mean_squared_error(Y_test,xgb_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(Y_test,xgb_pred)
print(f'Mean Absolute Error: {mae}')
r2 = r2_score(Y_test, xgb_pred)
print(f'R-squared: {r2:4f}')
```
